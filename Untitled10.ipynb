````python
import time
import random
from nba_api.stats.endpoints import leaguedashplayerstats, scoreboardv3

def safe_request(api_call_function, *args, **kwargs):
    """
    Wrapper to prevent API blocks.
    Includes a random delay between 0.6 and 1.5 seconds.
    """
    time.sleep(random.uniform(0.6, 1.5))
    try:
        return api_call_function(*args, **kwargs)
    except Exception as e:
        print(f"Connection Error: {e}. Retrying in 5 seconds...")
        time.sleep(5)
        return api_call_function(*args, **kwargs)

def update_intelligence_hub():
    # 1. Get All Active Teams & Conferences
    nba_teams = teams.get_teams()
    west = ['Nuggets', 'Warriors', 'Lakers', 'Suns', 'Mavericks', 'Clippers',
            'Kings', 'Pelicans', 'Thunder', 'Timberwolves', 'Rockets', 'Jazz',
            'Grizzlies', 'Trail Blazers', 'Spurs']

    team_data = []
    for t in nba_teams:
        conf = 'Western' if t['nickname'] in west else 'Eastern'
        team_data.append([t['id'], t['full_name'], t['abbreviation'], conf])

    # 2. Get Advanced Stats for ALL Active Players
    print("Fetching Advanced Player Metrics...")
    advanced_stats = safe_request(leaguedashplayerstats.LeagueDashPlayerStats,
                                  season='2023-24',
                                  measure_type_detailed_defense='Advanced')

    df_advanced = advanced_stats.get_data_frames()[0]

    # 3. Daily Game Scan
    today = datetime.now().strftime('%Y-%m-%d')
    print(f"Scanning games for {today}...")
    sb = safe_request(scoreboardv3.ScoreboardV3, game_date=today)
    df_games = sb.game_header.get_data_frame()

    # Debugging: Print columns to identify correct names
    print(f"Available columns: {df_games.columns.tolist()}")

    # --- SYNC TO GOOGLE SHEETS ---
    def get_or_create_worksheet(title):
        try:
            return sh.worksheet(title)
        except:
            return sh.add_worksheet(title=title, rows='100', cols='20')

    # Update Teams
    ws_teams = get_or_create_worksheet('Teams')
    ws_teams.clear()
    ws_teams.update([['ID', 'Name', 'Abbr', 'Conference']] + team_data, 'A1')

    # Update Player Metrics
    ws_stats = get_or_create_worksheet('Player_Stats')
    ws_stats.clear()
    ws_stats.update([df_advanced.columns.values.tolist()] + df_advanced.values.tolist())

    # Update Games
    ws_games = get_or_create_worksheet('Today_Games')
    ws_games.clear()
    if not df_games.empty:
        # Attempting to use universal keys if present, otherwise using what we find
        cols_to_use = [c for c in ['gameId', 'homeTeamId', 'awayTeamId', 'gameStatusText'] if c in df_games.columns]
        if len(cols_to_use) < 4:
             # Fallback to common variations
             mapping = {'gameId': 'gameId', 'homeTeamId': 'homeTeamId', 'awayTeamId': 'awayTeamId', 'gameStatusText': 'gameStatusText'}
             # This will help us identify if they are just capitalized differently
             for col in df_games.columns:
                 for key in mapping:
                     if col.lower() == key.lower():
                         cols_to_use.append(col)

        summary = df_games[list(set(cols_to_use))]
        ws_games.update([summary.columns.values.tolist()] + summary.values.tolist())
        print(f"✓ Success: {len(df_games)} games synced.")
    else:
        print('No games today.')

# Execute
update_intelligence_hub()
````

````python
from nba_api.stats.endpoints import leaguedashptstats, leaguedashteamstats, leaguedashplayerstats

def calculate_mismatches():
    print("🤖 Starting Intelligence Cross-Check...")

    # 1. Get Team Defensive Rankings (Opponent Strength)
    team_stats = safe_request(leaguedashteamstats.LeagueDashTeamStats, measure_type_detailed_defense='Advanced')
    df_teams = team_stats.get_data_frames()[0][['TEAM_ID', 'TEAM_NAME', 'DEF_RATING']]
    # Sort teams by Defensive Rating (Higher = Worse Defense)
    df_teams['DEF_RANK'] = df_teams['DEF_RATING'].rank(ascending=False)

    # 2. Get Player "Recency" Data (Last 10 Games)
    print("Analyzing Player Form (L10 vs Season)...")
    l10_stats = safe_request(leaguedashplayerstats.LeagueDashPlayerStats, last_n_games=10)
    df_l10 = l10_stats.get_data_frames()[0][['PLAYER_ID', 'PLAYER_NAME', 'PTS', 'REB', 'AST', 'FG3M']]

    season_stats = safe_request(leaguedashplayerstats.LeagueDashPlayerStats)
    df_season = season_stats.get_data_frames()[0][['PLAYER_ID', 'PTS', 'REB', 'AST']]

    # 3. Join Form Data
    analysis = df_l10.merge(df_season, on='PLAYER_ID', suffixes=('_L10', '_Season'))

    # 4. Logic: Identify "Hot" Players (Scoring 15% more than their season average)
    analysis['HOT_STREAK'] = analysis['PTS_L10'] > (analysis['PTS_Season'] * 1.15)

    # 5. Cross-Check with Today's Schedule
    try:
        ws_today = sh.worksheet('Today_Games')
        today_games_data = ws_today.get_all_records()
        if not today_games_data:
            print("No games found in the 'Today_Games' sheet.")
            return
        today_games = pd.DataFrame(today_games_data)
    except Exception as e:
        print(f"Could not access 'Today_Games' sheet: {e}")
        return

    # Sync Results to a 'Predictions' Sheet
    def get_or_create_worksheet(title):
        try:
            return sh.worksheet(title)
        except:
            return sh.add_worksheet(title=title, rows='1000', cols='20')

    ws_preds = get_or_create_worksheet('Predictions')
    ws_preds.clear()
    ws_preds.update([analysis.columns.values.tolist()] + analysis.values.tolist(), 'A1')
    print("✓ Mismatch Analysis Complete. Check the 'Predictions' sheet.")

# Execute
calculate_mismatches()
````

````python
import pandas as pd
import gspread
import os
import json
import time
import random
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguedashplayerstats, scoreboardv3, leaguedashteamstats
from google.oauth2.service_account import Credentials

# --- AUTHENTICATION ---
def get_gspread_client():
    scope = ['https://www.googleapis.com/auth/spreadsheets', 'https://www.googleapis.com/auth/drive']
    secret_json = os.environ.get('GCP_SERVICE_ACCOUNT')
    if not secret_json:
        raise ValueError("GCP_SERVICE_ACCOUNT environment variable not set.")
    service_account_info = json.loads(secret_json)
    creds = Credentials.from_service_account_info(service_account_info, scopes=scope)
    return gspread.authorize(creds)

# --- API WRAPPER ---
def safe_request(api_call_function, *args, **kwargs):
    time.sleep(random.uniform(1.0, 2.0))
    try:
        return api_call_function(*args, **kwargs)
    except Exception as e:
        print(f"Connection Error: {e}. Retrying...")
        time.sleep(5)
        return api_call_function(*args, **kwargs)

# --- BOT LOGIC ---
def run_pipeline():
    gc = get_gspread_client()
    sh = gc.open('NBA_Intelligence_Hub')

    print("Starting NBA Intelligence Pipeline...")

    # 1. Update Stats & Games
    today = datetime.now().strftime('%Y-%m-%d')
    sb = safe_request(scoreboardv3.ScoreboardV3, game_date=today)
    df_games = sb.game_header.get_data_frame()

    ws_games = sh.worksheet('Today_Games')
    ws_games.clear()
    if not df_games.empty:
        cols = [c for c in ['gameId', 'homeTeamId', 'awayTeamId', 'gameStatusText'] if c in df_games.columns]
        ws_games.update(range_name='A1', values=[cols] + df_games[cols].values.tolist())

    # 2. Calculate Mismatches
    l10_stats = safe_request(leaguedashplayerstats.LeagueDashPlayerStats, last_n_games=10).get_data_frames()[0]
    season_stats = safe_request(leaguedashplayerstats.LeagueDashPlayerStats).get_data_frames()[0]

    analysis = l10_stats[['PLAYER_ID', 'PLAYER_NAME', 'PTS']].merge(
        season_stats[['PLAYER_ID', 'PTS']], on='PLAYER_ID', suffixes=('_L10', '_Season')
    )
    analysis['HOT_STREAK'] = analysis['PTS_L10'] > (analysis['PTS_Season'] * 1.15)

    ws_preds = sh.worksheet('Predictions')
    ws_preds.clear()
    ws_preds.update(range_name='A1', values=[analysis.columns.tolist()] + analysis.values.tolist())

    # 3. Final Report
    sh.worksheet('Predictions').update(range_name='A1', values=[[f"FINAL PICKS FOR {today}"]])
    print("Pipeline complete. Sheets updated.")

if __name__ == '__main__':
    run_pipeline()
````
